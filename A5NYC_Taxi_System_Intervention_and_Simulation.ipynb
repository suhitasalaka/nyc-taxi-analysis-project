## Act 5: Pressure on the System

In this phase, the analysis shifts from observation to intervention.

Objective:
- Simulate a policy change in the NYC taxi system
- Evaluate how the system responds to the change
- Determine whether the system absorbs, amplifies, or reacts unpredictably

The magnitude of change is based on R = 4.

## Parameter Selection

We simulate a policy where base fare increases by a factor tied to R.

Justification:
Fare is a central economic variable influencing both driver and passenger behavior.
A small increase may affect tipping, trip frequency, and overall system efficiency.

Chosen change:
Fare increase = R × 2% = 4 x 2% = 8%

```python
import pandas as pd

df = pd.read_csv("clean_taxi_data.csv")

df.head()
```

```python
#cleaning tip% 
# # Ensure no zero or negative fares
df = df[df['fare_amount'] > 0]

# Recalculate tip percentage safely
df['tip_percentage'] = (df['tip_amount'] / df['fare_amount']) * 100

# Remove invalid values
df = df.replace([float('inf'), -float('inf')], pd.NA)
df = df.dropna(subset=['tip_percentage'])
```

```python
#applying simulation
R = 4
increase_factor = 1 + (R * 0.02)   # 8% increase

df_sim = df.copy()

# Apply fare increase
df_sim['fare_amount_new'] = df_sim['fare_amount'] * increase_factor

# Recalculate tip %
df_sim['tip_percentage_new'] = (df_sim['tip_amount'] / df_sim['fare_amount_new']) * 100

# Clean again
df_sim = df_sim.replace([float('inf'), -float('inf')], pd.NA)
df_sim = df_sim.dropna(subset=['tip_percentage_new'])
```

```python
#comparing before vs after
def summarize(data, fare_col, tip_col, label):
    print(f"\n--- {label} ---")
    print("Avg Fare:", round(data[fare_col].mean(), 2))
    print("Avg Tip %:", round(data[tip_col].mean(), 2))

summarize(df, 'fare_amount', 'tip_percentage', "ORIGINAL")
summarize(df_sim, 'fare_amount_new', 'tip_percentage_new', "AFTER POLICY")
```

## System Response Analysis

We evaluate how the system reacts to the simulated fare increase.

Observations:
- Average fare increases as expected
- Tip percentage slightly decreases due to higher fare denominator
- Distribution of tipping behavior shifts marginally

Interpretation:
The system does not show extreme shifts, suggesting stability.

### 1.Does the system absorb the change?

Yes,The system adjusts without drastic changes in behavior.



### 2.Does it amplify the change?

No,The impact remains proportional and does not escalate.



### 3.Does it respond in unexpected ways?

Partially,Tip percentage decreases slightly, which may indicate sensitivity to fare changes.

## Final Conclusion

The system demonstrates resilience to moderate fare increases.

While the policy change affects key metrics such as tip percentage,
the overall system behavior remains stable.

This suggests that the NYC taxi ecosystem can absorb small economic changes
without significant disruption.

However, subtle behavioral shifts indicate that passengers may adjust their tipping behavior
in response to increased fares.

Thus, the system is stable but sensitive to economic parameters.
